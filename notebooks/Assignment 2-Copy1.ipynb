### Checkout BIC Scores

```python
from matplotlib import pyplot as plt
from adjustText import adjust_text

def plot_BIC_scores(BICs):
    plt.figure(figsize=(15,14))
    color_iter = ['red', 'purple', 'blue', 'green']
    texts = []
    for index, (key, value) in enumerate(BICs.items()):
        x = np.arange(2,len(value)+1)
        y = value[1:]
        steps = np.negative(np.diff(value))
        nor_steps = steps/1.5
        plt.plot(1, value[0], "x", label=key, alpha=0.4, color=color_iter[index])
        plt.errorbar(x, y, yerr=nor_steps, alpha=0.4, fmt="--x", elinewidth=3, color=color_iter[index])
        for x, y, step in zip(x,y,steps):
            texts.append(plt.text(x=x, y=y, s=str(round(step)), color=color_iter[index]))
    plt.xticks(np.arange(1, len(value)+1, 1))
    plt.legend(title='Covariance type')
    plt.title('BIC vs Covariance Type')
    plt.grid(True, which='both')
    adjust_text(texts, arrowprops=dict(arrowstyle="->", color='r')) # this may take some time to compute
    plt.show()

def plot_gradient_BIC_scores(BICs, height=10):
    fig = plt.figure(figsize=(15, height))
    color_iter = ['red', 'purple', 'blue', 'green']
    for index, (key, value) in enumerate(BICs.items()):
        plt.subplot(len(BICs), 1, index+1)
        x = np.arange(1, len(value)) # 19 x in range 2-20
        gradients = np.diff(value) # 19 gradients of 20 data points
        y = gradients[1:] # 18 steps
        steps = np.negative(np.diff(gradients)) # 18 stepsc
        plt.plot(1, gradients[0], "x", label=key, alpha=0.4, color=color_iter[index])
        plt.errorbar(x[1:], y, yerr=steps, alpha=0.4, fmt="-x", elinewidth=3, color=color_iter[index])
        for _x, _y, step in zip(x[1:],y,steps):
            plt.text(x=_x*1.02, y=_y+(1.03*max(abs(y))), s=str(round(step)), color="black", fontsize=11)
        for _x, _y, gradient in zip(x,gradients,gradients):
            plt.text(x=_x*1.02, y=_y-(0.3*max(abs(y))), s=str(round(gradient)), color=color_iter[index], fontsize=11)
        plt.xticks(np.arange(1, len(value)+1, 1))
        plt.grid(True, which='both')
        plt.legend(title='Covariance type',loc=2)
    fig.suptitle('Gradinet of BIC Between Covariance Type')
    plt.show()
```

```python
GMM_intensity_full_9 = GMMs_intensity.get("full")[9]

def get_GMM_labels(input_x, GMM):
    # get predicted label for each data point
    labels = GMM.predict(input_x)
    return labels

def get_predict_proba(input_x, GMM):
    # get only the highest probability which is the predicted label probability
    return GMM.predict_proba(input_x).max(axis=1)

labels = get_GMM_labels(single_person_intensity_AUs, GMM_intensity_full_9)
probs = get_predict_proba(single_person_intensity_AUs, GMM_intensity_full_9)

# make a new copy dataframe
single_person_au_intensity_labeled_prob = single_person_au_intensity_preprocessed.copy(deep=True)
single_person_au_intensity_labeled_prob["label"] = labels
single_person_au_intensity_labeled_prob["prob"] = probs

reset_index = single_person_au_intensity_labeled_prob.reset_index()
sorted_reset_index = reset_index.sort_values(["prob", "label"],ascending=False)[["file","frame","prob", "label"]]

# extract 6 highest probability samples for each label
clc = {}
number_samples = 6
for label in labels:
    records = []
    files = []
    for row in sorted_reset_index[sorted_reset_index["label"] == label].itertuples():
        if len(records) < number_samples and row.file not in files:
            files.append(row.file)
            records.append(row)
        elif len(records) >= number_samples:
            break
    clc.update({label: pd.DataFrame(records).set_index("Index")})
```

## Final Discussion

It is very clear that similar frames such as _frames_test/53/12.png_ and _frames_test/53/15.png_ are labelled differently. The _frames_test/53/15.png_ is labelled as "Happiness - Smiling, lip corner angled upwards, eyes wrinkled" which is fairly accurate, and the other is "Disgust - Lip corner turned downwards, nose wrinkled". The second one looks like a similar emotion expressed by Phoebe. This kind of failure happens across the whole experiment. If the detections of AUs are fairly accurate. I guess there might be some mistakes in my work such as overfitting or wrong judgement on selecting the cluster size or etc. Also, it could be the limitation of GMM and BIC score. I think in a real research environment, one model is far away from enough for a clustering job like this. Also, using one cluster performance evaluation tool on trained model scores is not enough as well. It is better to use AIC, Silhouette Scores and other tools together and try to find the overall best choice. 

The dataset provided is limited on one person and it is not a "wild" dataset. It may contain differences in emotions when people are acting it and people are naturally expressing it. Also, it is better to have both males and females and multiple people in the dataset.

I guess it will be biased to use this GMM to classify other people's facial expressions. The dataset is very limited. Other people's facial expression could be different to Phoebe's in terms of activation of facial action units.
